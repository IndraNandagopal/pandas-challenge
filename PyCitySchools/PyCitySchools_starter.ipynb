# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(20)
```

## District Summary

```python
# Calculate the Totals (Schools and Students)
school_count = len(school_data_complete["school_name"].unique())
student_count = school_data_complete["Student ID"].count()
      
# Calculate the Total Budget
total_budget = school_data["budget"].sum()
```

```python
# Calculate the Average Scores
average_math_score = school_data_complete["math_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()
```

```python
school_data_complete.dtypes
```

```python
# Calculate the Percentage Pass Rates
math_pass_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count() ["math_score"]
math_pass_rate = (math_pass_count / float(student_count)) * 100
reading_pass_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count() ["reading_score"]
reading_pass_rate = (reading_pass_count / float(student_count)) * 100
math_reading_pass_count = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count() ["Student ID"]
#reading_pass_rate = (reading_pass_count / float(student_count)) * 100
math_reading_pass_rate = (math_reading_pass_count / float(student_count)) * 100
print(f'"Math Pass Rate:" {math_pass_rate:.6f}, "Reading Pass Rate:" {reading_pass_rate:.6f}, "Math Reading Pass Rate:" {math_reading_pass_rate:.6f}')

```

```python
# Minor Data Cleanup
district_summary = pd.DataFrame ({"Total Schools": [school_count],
                                "Total Students": [student_count],
                                "Total Budget": [total_budget],
                                "Average Math Score": [average_math_score],
                                "Average Reading Score": [average_reading_score],
                                "% Passing Math": [math_pass_rate],
                                "% Passing Reading":[reading_pass_rate],
                                "% Overall Passing": [math_reading_pass_rate]})


# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Determine the School Type
school_type = school_data.set_index(["school_name"])["type"]

# Calculate the total student count
schoolwise_student_count = school_data_complete["school_name"].value_counts()

# Calculate the total school budget and per capita spending
# per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
total_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_capita_spending = total_school_budget / schoolwise_student_count

# Calculate the average test scores
school_avg_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]
school_avg_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Get the students who passed math and passed reading by creating separate filtered DataFrames.
students_passed_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
students_passed_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]

# Get the the students who passed both reading and math in a separate DataFrame.
students_passed_math_and_reading = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
```

```python
#  Calculate the Percentage Pass Rates
schoolwise_math_pass_rate = students_passed_math.groupby(["school_name"]).count()["student_name"] / schoolwise_student_count * 100
schoolwise_reading_pass_rate = students_passed_reading.groupby(["school_name"]).count()["student_name"] / schoolwise_student_count * 100
schoolwise_overall_pass_rate = students_passed_math_and_reading.groupby(["school_name"]).count()["student_name"] / schoolwise_student_count * 100
```

```python
# Convert to DataFrame
per_school_summary = pd.DataFrame({"School Type": school_type,
                                   "Total Students": schoolwise_student_count,
                                   "Total School Budget": total_school_budget,
                                   "Per Student Budget": per_capita_spending,
                                   "Average Math Score": school_avg_math_score,
                                   "Average Reading Score": school_avg_reading_score,
                                   "% Passing Math": schoolwise_math_pass_rate,
                                   "% Passing Reading": schoolwise_reading_pass_rate,
                                   "% Overall Passing": schoolwise_overall_pass_rate})

# Minor data wrangling
per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,}".format)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:.2f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:.2f}".format)
per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:.2f}".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:.2f}".format)
per_school_summary["% Overall Passing"] = per_school_summary["% Overall Passing"].map("{:.2f}".format)

# Display the DataFrame
per_school_summary
```
